```python
from torchvision import datasets
import torch
```

```python
data_dir = "../datasets/mnist"

train_data = datasets.MNIST(data_dir, train=True, download=True)
```

```python
train_images = train_data.data
train_labels = train_data.targets

print(train_images.shape)
print(train_labels.shape)
```

```python
unique_values = train_labels.unique()
print(unique_values)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

R, C = len(unique_values), 10
fig, ax = plt.subplots(R, C, figsize=(10, 10))

for label_class, plot_row in enumerate(ax):
    label_x_rows = np.where(train_labels == label_class)[0]
    for plot_col in plot_row:
        plot_col.grid(False); plot_col.axis("off")
        ix = np.random.choice(label_x_rows)
        x, y = train_images[ix], train_labels[ix]
        plot_col.imshow(x, cmap="gray")
        plot_col.set_title(f"Label: {y}")

plt.tight_layout()
plt.show()
```
